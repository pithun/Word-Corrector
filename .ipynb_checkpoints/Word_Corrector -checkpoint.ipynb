## Creating the combinations algorithm. COMBOBOOST

```python
# below i tried to slice the words with probabilities higher than the 98th percentile probabilities for 
# the first 8 word in the test set. Basically random forest assigns probability to each word in the train
# set what i did below is for the first misspelt word 'gypt', i printed out the words with highest 
# probability of being the correct word for 'gypt'
index = []
for c in range(8):
    for ind,rand in enumerate(rand_pred[c]):
        if rand >= np.percentile(rand_pred[c], 98):
            index.append(ind)
    print(y_test(dataframe, index, 0))
    index = []
    # gypt', 'rusia', 'gemany', 'canda', 'nygeria', 'borkinafaso', 'anada', 'states']
# it actually works the way it should
```

```python
# basically to use my idea i had to know what probability to take the probabilities of my model from and
# i've seen that most times the right answer should be above the 98th percentile.
for per in range(8):
    cdf = empiricaldist.Cdf.from_seq(rand_pred[per])
    print(cdf.inverse(0.98))
    cdf.plot(linestyle= 'None', marker= '.', grid= True)
    plt.show()
# let's check this out for Nigeria that was predicted as germany basically i'm saying that nigerias pr 
# should be high enough to make the 98th prcentile probability. 
```

```python
def looper (onerawperm, predicted_permutes):
    count = 0
    percentages = []
    for a in predicted_permutes:
        #print(a)
        for b in onerawperm:
            if b in a:
                count+=1
        #print(count)
        percentages.append(((count/len(a))*100)*(count/len(b)))
        count= 0
    return percentages
```

```python
# basically after prediction what i want to do is get the >= 98th percentile probabilities, get their index
# in the y train and use combinations to get new probability and make a selection.
def comboboost(dataframe, raw_X_train, y_train, raw_test_data, col_to_get_from, base_model= 'MultinomialNB()', 
               combinations_to_subset = 4, percentile = 95):
    import pandas as pd
    import numpy as np
    import itertools
    from sklearn.naive_bayes import MultinomialNB
    from sklearn.feature_extraction.text import CountVectorizer
    model= MultinomialNB()
    count_vector = CountVectorizer(analyzer= 'char')
    char_array_train = count_vector.fit_transform(raw_X_train).toarray()
    vectorized_test_data = count_vector.transform(raw_test_data)
    frequency_matrix = pd.DataFrame(char_array_train, columns= count_vector.get_feature_names())
    frequency_matrix

    model.fit(frequency_matrix, y_train)
    index = []
    count = 0
    test = []
    actual_words = []
    # first i got combinations of the raw data in a list of lists
    permuted_raw_data = [list(itertools.combinations(d, combinations_to_subset)) for d in raw_test_data]
    
    
    # i got the probability list that random forest gave to all 254 output categories. e.g for ygypt
    # we'd have probabilities that the country is united states but a very low one. since our model is
    # basically an overfiting model.
    model_prob_predict = model.predict_proba(vectorized_test_data)

    # using the for loop and the number or rows in our prediction which is basically the number of test 
    # data for each row of prediction i got the 95th percentile probabilities basically words that are
    # most likely the word i want for each test misspelt word.
    for c in range(model_prob_predict.shape[0]):
        for ind,rand in zip(y_train, model_prob_predict[c]):
            if rand >= np.percentile(model_prob_predict[c], percentile):
                index.append(ind)
            else:
                pass
        # after getting the index of the most likely words i want, i used the y test fuction to 
        # actually get those words using the index
        
        test = [list(itertools.combinations(aa, combinations_to_subset)) for aa in y_test(dataframe, index, 
                                                                                          col_to_get_from)]
        # i created the looper function above and it basically looks at the word we are trying to 
        # predict breaks it into combinations and looks for those words in a list containing our
        # possible words also broken into combinations, the assumption is that the correct work will
        # have more combinations in common to the real word. so we assign probability based on number
        # of similar combinations and return those percentages, our real word is taken to be the one
        # with the highest probability.
        
        percentage = looper(permuted_raw_data[c], test)
        max_index = percentage.index(max(percentage))
        actual_words.append(index[max_index])
        index = []
        
    return y_test(dataframe, actual_words, col_to_get_from)
```

# Problem enountered.
Initial problem with this boost is that for example it's predicting coulumbia as cuba, basically, cuba has just one four word combination basically 4 C 4 = 1 so just one, now the idea of the algo is that we get combinations of the mistake and combinations of one of the possible correct words then we check how many combinations of the mistake we see in the correct word and divide by the no of combinations the correct word has in our case we can find cuba twice in coulumbia and just once in cuba so i.e 2/1 x 100 giving it 200. 

That's wrong to correct this i did a double comparison i.e count/ len(combinations in cuba) and count/len(combinations in coulumbia) the idea is that for the right word colombia, we'll have more in common so count/len(combinations in coulumbia) won't be small but for cuba count/ len(combinations in coulumbia) will be soo small and will make the metric soo small hence our real word will have a high weight and it works marvelously.
